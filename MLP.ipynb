```python
from sklearn.neural_network import MLPClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
df = pd.read_csv("car.csv")
df
```

```python
df.c1 = pd.Categorical(pd.factorize(df.c1)[0])
df.c2 = pd.Categorical(pd.factorize(df.c2)[0])
df.c3 = pd.Categorical(pd.factorize(df.c3)[0])
df.c4 = pd.Categorical(pd.factorize(df.c4)[0])
df.c5 = pd.Categorical(pd.factorize(df.c5)[0])
df.c6 = pd.Categorical(pd.factorize(df.c6)[0])
# df.Classification = pd.Categorical(pd.factorize(df.Classification)[0])
df
```

```python
X = df.drop(columns=['Classification'])
y = df.Classification 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)
clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(15), random_state=1)

clf.fit(X, y)
p=clf.predict(X_test)

iweight=[]
for i in range(len(list(X))):
    iweight+=[clf.coefs_[0][i][0]]
fweight=[]
for i in range(len(list(X))):
    fweight+=[clf.coefs_[0][i][-1]]
print("Initial Weights:")
print(iweight)
print("Final Weights:")
print(fweight)
```

```python
ib=clf.intercepts_[0][0]
fb=clf.intercepts_[0][-1]
print("Initial Bias: \n",ib,"\nFinal Bias \n",fb)
```

```python
print("------------------------")
print('Precision Score : ', precision_score(y_test, p, average= 'weighted', labels=np.unique(p)))
print('Recall Score \t: ', recall_score(y_test, p, average= 'weighted', labels=np.unique(p)))
acc = accuracy_score(y_test, p)
print('Accuracy \t: ', acc)
print('Error \t\t: ', 1-acc)
print("------------------------")
```
